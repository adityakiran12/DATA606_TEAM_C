```python
import yfinance as yf
import pandas as pd

# Define the ticker symbol
ticker_symbol = "AAPL"

# Define the date range
start_date = "2023-01-01"
end_date = "2024-02-14"

# Fetch the historical data
apple_stock_data = yf.download(ticker_symbol, start=start_date, end=end_date)

# Convert the fetched data to a DataFrame
apple = pd.DataFrame(apple_stock_data)

# Display the DataFrame
apple.head()
```

```python
from matplotlib import pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(apple.index, apple['Close'], color='blue')  # Replace 'apple['Date']' with 'apple.index'
plt.title('AAPL Closing Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Closing Price ($)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm

decomposition = sm.tsa.seasonal_decompose(apple['Close'], model='additive', period=30)

# Plot the decomposed components
plt.figure(figsize=(10, 8))

plt.subplot(411)
plt.plot(apple['Close'], label='Original', color='blue')
plt.legend(loc='upper left')

plt.subplot(412)
plt.plot(decomposition.trend, label='Trend', color='red')
plt.legend(loc='upper left')

plt.subplot(413)
plt.plot(decomposition.seasonal, label='Seasonality', color='green')
plt.legend(loc='upper left')

plt.subplot(414)
plt.plot(decomposition.resid, label='Residuals', color='purple')
plt.legend(loc='upper left')


plt.tight_layout()
plt.show()
```

```python
df_tesla_final['date'] = pd.to_datetime(df_tesla_final['date'])
```

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Assuming df_tesla_final is your DataFrame

# Compute the correlation matrix
corr = df_tesla_final.corr()

# Create a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
plt.figure(figsize=(10, 8))

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap='coolwarm', annot=True, fmt=".2f", linewidths=0.5)

plt.title('Correlation Matrix')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.cm import ScalarMappable
from sklearn.preprocessing import RobustScaler

# Convert 'Date' column to datetime
df_tesla_final['date'] = pd.to_datetime(df_tesla_final['date'])

# Create a new column for 'Next Day Open-Close'
df_tesla_final['Next Day Open-Close'] = df_tesla_final['Next Day Open'] - df_tesla_final['Close']

# Define the columns for plotting
columns_to_plot = ['Close', 'Open', 'Next Day Open-Close', 'Volume', 'count_positive', 'count_negative', 'count_neutral']

# Plotting
plt.figure(figsize=(12, 16))

for i, column in enumerate(columns_to_plot, 1):
    plt.subplot(7, 1, i)

    # Clip extreme values to 1st and 99th percentiles
    clipped_values = df_tesla_final[column].clip(df_tesla_final[column].quantile(0.01), df_tesla_final[column].quantile(0.99))

    # Scale the values using RobustScaler
    scaler = RobustScaler()
    scaled_values = scaler.fit_transform(clipped_values.values.reshape(-1, 1))

    # Create colormap based on scaled values
    norm = plt.Normalize(scaled_values.min(), scaled_values.max())
    cmap = plt.get_cmap('coolwarm')

    # Create color mapping for bars
    colors = cmap(norm(scaled_values))

    # Plot bars with color mapping
    plt.bar(df_tesla_final['date'], df_tesla_final[column], color=colors)

    plt.title(column)
    plt.xlabel('Date')
    plt.ylabel(column)

    # Add color bar to show mapping
    sm = ScalarMappable(cmap=cmap, norm=norm)
    sm.set_array([])
    plt.colorbar(sm, ax=plt.gca(), orientation='vertical')

plt.tight_layout()
plt.show()
```
